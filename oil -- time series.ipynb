```python
mean=np.round(df['dcoilwtico'].mean(),1)
```

```python
ts=df['dcoilwtico']
plt.plot(ts)
```

```python
def test_stationary(timeseries):
    rollmean=timeseries.rolling(window=12).mean()
    rollstd = timeseries.rolling(window=12).std()
    orig = plt.plot(timeseries,color='blue',label='Original')
    mean=plt.plot(rollmean,label='mean',color='black')
    std=plt.plot(rollstd,label='std',color='blue')
    plt.title('Rolling')
    plt.show()
    
    #dickey fuller test
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
test_stationary(ts)
```

### transformation

```python
ts_log=np.log(ts)
plt.plot(ts_log)
```

### differencing

```python
moving_average=ts_log.rolling(window=12).mean()
moving_std=ts_log.rolling(window=12).std()
ts_log_diff=ts_log-moving_average
ts_log_diff
```

### dropping null value

```python
ts_log_diff.dropna(inplace=True)
```

```python
test_stationary(ts_log_diff)
```

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
plt.plot(range(0,15),acf(ts_log_diff,nlags=14))
plt.axhline(y=0.2,linestyle='--')
plt.axhline(y=0.4,linestyle='--')
plt.axhline(y=0,linestyle='--')
plt.title('Autocorrelation')
```

### Pacf

```python
plt.plot(range(0,15),pacf(ts_log_diff,nlags=14))
plt.axhline(y=0.2,linestyle='--')
plt.axhline(y=0.4,linestyle='--')
plt.axhline(y=0,linestyle='--')
plt.title('Partial autocorrelation')
```

```python
from statsmodels.tsa.arima_model import ARIMA,ARMA
```

```python
model=ARMA(ts_log_diff,order=(1,1)).fit()
plt.title('ARMA')
plt.plot(ts_log_diff)
plt.plot(model.fittedvalues,color='r')
```

```python
model=ARIMA(ts_log_diff,order=(1,1,1)).fit()
plt.title('ARMA')
plt.plot(ts_log_diff)
plt.plot(model.fittedvalues,color='r')
```
